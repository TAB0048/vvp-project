# Incidence matrix
Examples of creating incidence matrices for a maze.

```python
incidence_matrix1 = maze1.incidence_matrix()
print(incidence_matrix1.todense())
```

```python
incidence_matrix4 = maze4.incidence_matrix()
print(incidence_matrix4.todense())
```

```python
incidence_matrix6 = maze6.incidence_matrix()
print(incidence_matrix6.todense())
```

# Dijkstra's algorithm
Examples of performing Dijkstra's algorithm.

```python
print(maze1.dijkstra(incidence_matrix1))
```

```python
print(maze4.dijkstra(incidence_matrix4))
```

```python
print(maze6.dijkstra(incidence_matrix6))
```

# Maze templates
Examples of available maze templates and their solutions.

```python
empty_maze = MazeTemplate.empty(10)
empty_maze.draw_maze()
empty_maze.draw_maze_path()
```

```python
slalom_maze = MazeTemplate.slalom(50)
slalom_maze.draw_maze()
slalom_maze.draw_maze_path()
```

```python
lines_maze = MazeTemplate.lines(60)
lines_maze.draw_maze()
lines_maze.draw_maze_path()
```

# Maze templates with obstacles

```python
num_obstacles = 10
empty_maze_obstacles = MazeTemplate.random_obstacles(empty_maze, num_obstacles)

empty_maze_obstacles.draw_maze()
empty_maze_obstacles.draw_maze_path()
```

```python
num_obstacles = 80
slalom_maze_obstacles = MazeTemplate.random_obstacles(slalom_maze, num_obstacles)

slalom_maze_obstacles.draw_maze()
slalom_maze_obstacles.draw_maze_path()
```

```python
num_obstacles = 2
lines_maze_obstacles = MazeTemplate.random_obstacles(lines_maze, num_obstacles)

lines_maze_obstacles.draw_maze()
lines_maze_obstacles.draw_maze_path()
```

Example of saving a created maze as a new CSV file.
